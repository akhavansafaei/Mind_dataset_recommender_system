```python
import numpy as np

# Function to extract features from the dictionary
def get_news_features(news_ids, news_dict):
    features = [
        news_dict[news_id][:768]  # Only take the first 768 elements
        for news_id in news_ids
        if news_id in news_dict
    ]
    if features:
        # Compute mean pooling
        return np.mean(np.array(features), axis=0)
    else:
        # Return zeros with size 768 if no valid news ID is found
        return np.zeros(768)

# Example: Adding user features to the DataFrame
user_df['user_features'] = user_df['explored_news'].apply(
    lambda x: get_news_features(x.split(), news_dict)  # Split `history` into news IDs
)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from tqdm import tqdm  # For progress bar

# Assuming `user_df['user_features']` contains 768-dimensional features
user_features = np.stack(user_df['user_features'])

# Range for clusters
k_values = range(2, 21)

# Lists to store evaluation metrics
silhouette_scores = []
sse = []  # Sum of squared errors for elbow method

# Evaluate clusters with progress bar
for k in tqdm(k_values, desc="Evaluating Clusters"):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(user_features)

    # Compute metrics
    sse.append(kmeans.inertia_)  # SSE
    silhouette_scores.append(silhouette_score(user_features, cluster_labels))

# Plotting
plt.figure(figsize=(12, 5))

# Elbow Method
plt.subplot(1, 2, 1)
plt.plot(k_values, sse, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')

# Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```

```python

# Function to find the most common news explored within each cluster
def get_common_news(cluster_data, top_n=50):
    # Combine all the news explored by users in the cluster
    all_explored_news = ' '.join(cluster_data['explored_news'].dropna())
    # Split the string into individual news IDs
    news_list = all_explored_news.split()
    # Count the occurrences of each news ID
    news_counts = Counter(news_list)
    # Get the top N most common news items
    common_news = news_counts.most_common(top_n)
    return set([news_id for news_id, _ in common_news])

# Get the top 50 common news for each cluster
top_50_common_news_by_cluster = {}
for cluster_id in range(k):
    cluster_data = user_df[user_df['cluster'] == cluster_id]
    top_50_common_news = get_common_news(cluster_data, top_n=50)
    top_50_common_news_by_cluster[cluster_id] = top_50_common_news

# Function to highlight text in blue and underlined
def highlight_common_news(news, common_news, top_50_news, user_common_news):
    highlighted_news = []
    for news_id in news.split():
        if news_id in common_news:
            if news_id in top_50_news:
                highlighted_news.append(f"\033[4m\033[34m{news_id}\033[0m")  # Underlined and blue for top 50
            elif news_id in user_common_news:
                highlighted_news.append(f"\033[4m{news_id}\033[0m")  # Underlined for common between users
            else:
                highlighted_news.append(f"\033[34m{news_id}\033[0m")  # Blue for common news
        else:
            highlighted_news.append(news_id)
    return ' '.join(highlighted_news)

# Select top 3 clusters (based on cluster_id) for analysis
top_clusters = sorted(top_50_common_news_by_cluster.keys())[:3]

# Print the results for each cluster
for cluster_id in top_clusters:
    cluster_data = user_df[user_df['cluster'] == cluster_id]
    common_news = get_common_news(cluster_data)
    top_50_news = top_50_common_news_by_cluster[cluster_id]

    # Select 10 random users who have explored fewer than 50 news
    eligible_users = cluster_data[cluster_data['explored_news'].apply(lambda x: len(x.split()) < 50)]
    selected_users = eligible_users.sample(n=10, random_state=42)

    print(f"\nCluster {cluster_id} - Example Users (highlighting common/top 50 news):")

    # Collect common news between the selected users
    selected_users_news = set()
    for _, row in selected_users.iterrows():
        selected_users_news.update(row['explored_news'].split())

    for _, row in selected_users.iterrows():
        explored_news = row['explored_news']
        # Highlight the news for this user
        highlighted_news = highlight_common_news(explored_news, common_news, top_50_news, selected_users_news)
        print(f"  User {row['user_id']} explored news: {highlighted_news}")
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Assuming `user_df['user_features']` contains 768-dimensional features and `user_df['explored_news']` contains the list of news IDs
user_features = np.stack(user_df['user_features'])

# Perform K-Means with 10 clusters
k = 10
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model and get cluster labels
cluster_labels = kmeans.fit_predict(user_features)

# Add the cluster labels to the user dataframe
user_df['cluster'] = cluster_labels

# Select a cluster (e.g., cluster 0 for visualization)
cluster_id = 0  # You can change this to another cluster ID to visualize a different cluster
cluster_data = user_df[user_df['cluster'] == cluster_id]

# Initialize the graph
G = nx.Graph()

# Add nodes for users
for _, row in cluster_data.iterrows():
    user_node = f"User_{row['user_id']}"
    G.add_node(user_node, type='user')

    # Add edges between user and news items they explored
    explored_news = row['explored_news'].split()
    for news_id in explored_news:
        news_node = f"News_{news_id}"
        G.add_node(news_node, type='news')
        G.add_edge(user_node, news_node)

# Create a layout for the graph
pos = nx.spring_layout(G, k=0.15, iterations=20)

# Draw the graph
plt.figure(figsize=(12, 12))

# Draw user nodes (circle-shaped, blue)
user_nodes = [node for node, data in G.nodes(data=True) if data['type'] == 'user']
nx.draw_networkx_nodes(G, pos, nodelist=user_nodes, node_color='skyblue', node_size=500, label='Users')

# Draw news nodes (square-shaped, green)
news_nodes = [node for node, data in G.nodes(data=True) if data['type'] == 'news']
nx.draw_networkx_nodes(G, pos, nodelist=news_nodes, node_color='lightgreen', node_size=300, label='News')

# Draw edges (gray)
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.7, edge_color='gray')

# Draw labels for nodes (users and news)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')

# Show the plot
plt.title(f"Explored News Network for Cluster {cluster_id}")
plt.axis('off')  # Hide the axes
plt.show()
```
